```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Use a clean style
plt.style.use('seaborn-v0_8')
```

```python
category_sales = df.groupby('category')['sale_price'].sum().sort_values()

plt.figure(figsize=(8,5))
category_sales.plot(kind='barh', color=['#4C72B0','#DD8452','#55A868'])
plt.title('Total Sales by Category')
plt.xlabel('Total Sale Price ($)')
plt.tight_layout()
plt.savefig('assets/sales_by_category.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
df['order_date'] = pd.to_datetime(df['segment'].iloc[:,1] if 'order_date' not in df.columns else df['order_date'])
monthly = df.groupby(df['order_date'].dt.to_period('M'))['sale_price'].sum()

plt.figure(figsize=(14,5))
monthly.plot(kind='line', color='steelblue', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.savefig('assets/monthly_trend.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
region_sales = df.groupby('region')['sale_price'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.barplot(x=region_sales.index, y=region_sales.values, palette='coolwarm')
plt.title('Total Sales by Region')
plt.ylabel('Total Sales ($)')
plt.tight_layout()
plt.savefig('assets/sales_by_region.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
profit_cat = df.groupby('category')['profit'].sum().sort_values()

plt.figure(figsize=(8,5))
colors = ['green' if x > 0 else 'red' for x in profit_cat.values]
profit_cat.plot(kind='barh', color=colors)
plt.title('Total Profit by Category')
plt.xlabel('Total Profit ($)')
plt.tight_layout()
plt.savefig('assets/profit_by_category.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
sub_profit = df.groupby('sub_category')['profit'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=sub_profit.values, y=sub_profit.index, palette='viridis')
plt.title('Top 10 Most Profitable Sub-Categories')
plt.xlabel('Total Profit ($)')
plt.tight_layout()
plt.savefig('assets/top_subcategories.png', dpi=150, bbox_inches='tight')
plt.show()
```
